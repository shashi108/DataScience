```python
#%%%%%%%%%%%%%%%%%%%%%% Basic prompt template: simple, no-variable template %%%%%%%%%%%%%%%%%%%
from langchain_community.llms import Ollama
from langchain.prompts import PromptTemplate #Import PromptTemplate from LangChain
# Creating a simple prompt template
prompt_template = PromptTemplate.from_template("Tell me a joke")
# Formatting and displaying the prompt
prompt =prompt_template.format()
print(prompt)
```

```python
#%%% Variable-embedded prompt template: template with variables, allowing for dynamic content insertion %%%%%%%%%
# Using PromptTemplate for dynamic content
prompt_template = PromptTemplate.from_template("Tell me a {adjective} joke about {content}.")
# Inserting variables into the prompt
prompt=prompt_template.format(adjective="funny", content="chickens")
print(prompt)
```

```python
#pip install ctransformers
from langchain.prompts import PromptTemplate
from langchain_community.llms import CTransformers
# Creating a simple prompt template with placeholders
template = PromptTemplate(
                           template="Tell me a {adjective} joke about {content}.",
                           input_variables=["adjective", "content"]
                           )
# Call the Llama model using CTransformer
llm = CTransformers(model="TheBloke/Llama-2-7B-Chat-GGUF", model_type="llama")
# Formatting the prompt with specific inputs
prompt = template.format(adjective="funny", content="chickens")
# Generating a response from the language model
response = llm(prompt)
print(response)
```

```python
# Importing ChatPromptTemplate from LangChain_core
from langchain_core.prompts import ChatPromptTemplate
# Creating a chat prompt template with dialogue
chat_template = ChatPromptTemplate.from_messages([
                                                 ("system", "You are a helpful AI bot. Your name is Bob."),
                                                 ("human", "Hello, how are you doing?"),
                                                 ("ai", "I'm doing well, thanks!"),
                                                 ("human", "What can you tell me about Langchain?")])
# Formatting and printing each chat message
formatted_messages = chat_template.format_messages()
for message in formatted_messages:
    content = message.content if hasattr(message, 'content') else str(message)
    print(content)
```

```python
from langchain.prompts.few_shot import FewShotPromptTemplate
from langchain.prompts.prompt import PromptTemplate
from langchain_community.llms import CTransformers
# Define a smaller set of few-shot examples
examples = [
            {"question": "Question: Who was the first person to walk on the moon?",
              "answer": "Answer: Neil Armstrong"},
            {"question": "Question: What is the capital of France?",
              "answer": "Answer: Paris"}
            ]
# Configure a simple formatter for the few-shot
examples
example_prompt = PromptTemplate(
                                 input_variables=["question", "answer"],
                                 template="{question}: {answer}"
                                 )
# Create a FewShotPromptTemplate object with examples and formatter
prompt = FewShotPromptTemplate(examples=examples,example_prompt=example_prompt,suffix="{input}",input_variables=["input"])
# Format the prompt with a specific question
formatted_prompt = prompt.format(input="What isthe largest planet in our Solar System?")
llm = CTransformers(model="TheBloke/Llama-2-7B-Chat-GGUF", model_type="llama")
# Get a response from the language model
response = llm(formatted_prompt)
print(response)
```
