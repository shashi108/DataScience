# 2.Document splitting

##### How to split by HTML header
HTMLHeaderTextSplitter is a "structure-aware" chunker that splits text at the HTML element level and adds metadata for each header "relevant" to any given chunk. It can return chunks element by element or combine elements with the same metadata, with the objectives of (a) keeping related text grouped (more or less) semantically and (b) preserving context-rich information encoded in document structures. It can be used with other text splitters as part of a chunking pipeline.


```python
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://plato.stanford.edu/entries/goedel/"

# Extract Header 1,2,3,4
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
print(html_header_splits[0].page_content)
print("\n")
print(html_header_splits[1].page_content[:100])
```

#### How to split by character-Character Text Splitter
This is the simplest method. This splits based on a given character sequence, which defaults to "\n\n". Chunk length is measured by number of characters.

1. How the text is split: by single character separator.
2. How the chunk size is measured: by number of characters.

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size = 80,chunk_overlap = 5)
text_doc=text_splitter.create_documents([txt_doc]) 
print(text_doc)
splits[0].page_content
```

#### Text Splitting from Documents- RecursiveCharacter Text Splitters
This text splitter is the recommended one for generic text. It is parameterized by a list of characters. It tries to split on them in order until the chunks are small enough. The default list is ["\n\n", "\n", " ", ""]. This has the effect of trying to keep all paragraphs (and then sentences, and then words) together as long as possible, as those would generically seem to be the strongest semantically related pieces of text.

- How the text is split: by list of characters.
- How the chunk size is measured: by number of characters.

```python
# Document splitting
# Import the text splitter from LangChain
from langchain.text_splitter import RecursiveCharacterTextSplitter
# Initialize splitter with specified chunk size and overlap
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1500,chunk_overlap = 150)
splits = text_splitter.split_documents(docs_pdf) # Split the documents into chunks
print('Numbers of chunks:', len(splits)) # Print total number of chunks created
# Display metadata of the first split
print('Data associated with each split 0:', splits[0].page_content[:200])
print('\nData associated with each split 1:', splits[1].page_content[:200])
```

```python
#This function creates embeddings for a collection of documents
query_result=embeddings.embed_documents(splits)
query_result[:10]
```

```python
#This function typically generates an embedding for a single query or piece of text.
text="This is a tutorial on OPENAI embedding"
query_result_q=embeddings.embed_query(text)
query_result_q[:10]
```

#### Faiss
Facebook AI Similarity Search (Faiss) is a library for efficient similarity search and clustering of dense vectors. It contains algorithms that search in sets of vectors of any size, up to ones that possibly do not fit in RAM. It also contains supporting code for evaluation and parameter tuning.

```python
## building a sample vectordb
dbtype="FAISS"
if dbtype=="FAISS":
    from langchain_community.vectorstores import FAISS
    db=FAISS.from_documents(documents=splits,embedding=embeddings)
elif dbtype=="Chroma":
    from langchain_chroma import Chroma
    db=Chroma.from_documents(documents=splits,embedding=embeddings)
else:
    print("Select DB type")
```

```python
### querying 
query="Clinical study reports"
docs=db.similarity_search(query,k=2)
docs[0].page_content
```

#### Similarity Search with score
There are some FAISS specific methods. One of them is similarity_search_with_score, which allows you to return not only the documents but also the distance score of the query to them. The returned distance score is Euclidean (L2) distance. Therefore, a lower score is better.

```python
docs_and_score=db.similarity_search_with_score(query,k=2)
docs_and_score[0]
```

### Maximum marginal relevance search (MMR)
Maximum marginal relevance (MMR) focuses on relevance and diversity in search results:<br>

Relevance to the Query: selects items that are highly relevant to a given query or information need, ensuring that the chosen     items align closely with search intent or requirements.<br>

Diversity Among Results: presents varied items that collectively cover different aspects of the query, preventing redundancy and providing a more comprehensive view of the information sought by the user.<br>

```python
# Using MMR search criteria
docs_mmr=db.max_marginal_relevance_search(query, k=2, fetch_k=10)
docs_mmr[0].page_content
```

```python
embedding_vector=embeddings.embed_query(query)
embedding_vector[:10]
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
docs_score[0].page_content
```

### FAISS DB as a retriever
We can also convert the vectorstore into a Retriever class. This allows us to easily use it in other LangChain methods, which largely work with retrievers.

```python
retriever=db.as_retriever()
docs=retriever.invoke(query)
docs[0].page_content  # print first document content
```
